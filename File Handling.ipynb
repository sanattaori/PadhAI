```
x = 100
y = 1
for i in range(1,x):
  y *= i
print('factorial of ',x, 'is {}'.format(y))
```

```
math.factorial(x)
```

```
tic = time.time()
y = 1
for i in range(1,x):
  y *= i
print('factorial of ',x, 'is {}'.format(y))
toc = time.time()
own = toc - tic
print('time difference for own function ',own)
```

```
tic = time.time()
y = 1
print('factorial of {} is {} '.format(x,math.factorial(x)))
toc = time.time()
lib = toc - tic
print('time difference for package function ',lib)
```

```
own - lib
```
