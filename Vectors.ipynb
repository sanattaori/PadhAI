```
import numpy as np
import matplotlib.pyplot as plt
```

```
plt.quiver(0,0,3,4)
```

```
plt.quiver(0,0,3,4 , scale_units='xy', angles='xy',scale=1,color='r')
plt.quiver(0,0,-3,4 , scale_units='xy', angles='xy',scale=1,color='g')

plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```
def plot_vectors(vecs):
  colors = ['r','b','g']
  for vec in vecs:
    plt.quiver(0,0,3,4,scale_units='xy',angles='xy',scale=1,color='r')
  plt.xlim(-10,10)
  plt.ylim(-10,10)
  plt.show()
```

```
plot_vectors([(0,0,3,4),(0,0,-3,-4),(0,0,-3,4)])
```
