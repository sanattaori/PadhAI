# Plotting Sigmoid

```
import numpy as np
import matplotlib.pyplot as plt
```

```
def sigmoid(x,w,b):
  return 1/(1+ np.exp(-(w*x +b)))
```

```
sigmoid(1,.5,0)
```

```
w = -2  #@param {type: "slider", min: -2, max: 2}
b = 2 #@param {type: "slider", min: -2, max: 2}
x = np.linspace(-10,10,100)
Y = sigmoid(x,w,b)
```

```
plt.plot(x,Y)
plt.show()
```

```
plt.plot(x,sigmoid(x,-w,b))
```

```
plt.plot(x,sigmoid(x,-w,b=4))
```

```
plt.plot(x,sigmoid(x,-w,b=7))
```
