```
import numpy as np
```

```
def sigmoid(w,b,x):
	return 1.0/(1.0+np.exp(-(w*x+b)))
```

### Negative Slope

####W negative

For increasing value of x slope decreases

```
y = sigmoid(-2,0,x)
plt.plot(x,y)
```

```
y = sigmoid(-5,0,x)
plt.plot(x,y)
```

### Positive Slope

#### W positive

```
y = sigmoid(.5,0,x)
plt.plot(x,y)
```

```
y = sigmoid(4,0,x)
plt.plot(x,y)
```

### Decreasing value of b

```
y = sigmoid(1,4,x)
plt.plot(x,y)
```

```
y = sigmoid(1,-1,x)
plt.plot(x,y)
```

```
y = sigmoid(1,-5,x)
plt.plot(x,y)
```
